# Clay Data Exploration: Authentic Earth Observation
This notebook explores the datasets used for validating the Clay Foundation Model. We compare standard RGB satellite imagery with Synthetic Aperture Radar (SAR).

```
import torch
import numpy as np
import matplotlib.pyplot as plt
import torchvision
from PIL import Image
from datasets import load_dataset
```

## 1. EuroSAT RGB (Standard Visual Satellite)
Used for the primary benchmark visualization. Clean 3-channel imagery.

```
dataset_rgb = torchvision.datasets.EuroSAT(root='CLAY/data', download=True)
img, label = dataset_rgb[4] # Industrial example
plt.imshow(img)
plt.title(f'EuroSAT RGB - Class: Industrial')
plt.axis('off')
plt.show()
```

## 2. EuroSAT SAR (Sentinel-1 Radar)
SAR captures physical structure and moisture. It works through clouds and at night.

```
# Using the verified HuggingFace path for SAR
ds_sar = load_dataset('wangyi111/EuroSAT-SAR', split='train', streaming=True)
sample_sar = next(iter(ds_sar))
img_sar = np.array(sample_sar['image'])
print(f'SAR Shape: {img_sar.shape}')
```

```
# Visualize SAR VV and VH polarizations
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
vv = img_sar[0, :, :].astype(float)
vh = img_sar[1, :, :].astype(float)

ax1.imshow(vv, cmap='gray')
ax1.set_title('VV Polarization')
ax2.imshow(vh, cmap='gray')
ax2.set_title('VH Polarization')
plt.show()
```
