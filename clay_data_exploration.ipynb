# Clay Data Exploration: Earth Observation Datasets
This notebook explores the datasets used to train and validate the Clay Foundation Model.

```
import torch
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_dataset
from PIL import Image
```

## 1. EuroSAT MSI (Sentinel-2)
EuroSAT MSI consists of 27,000 labeled images with 13 spectral bands. Clay uses 10 of these bands (Sentinel-2 L2A format).

```
ds_msi = load_dataset('blanchon/EuroSAT_MSI', split='train', streaming=True)
sample = next(iter(ds_msi))
img_raw = np.array(sample['image'])
print(f'Shape: {img_raw.shape}')
print(f'Max Value: {img_raw.max()}') # Usually 16-bit
```

```
# Visualize RGB (Bands 4, 3, 2)
rgb = img_raw[[3, 2, 1], :, :].transpose(1, 2, 0).astype(float)
p2, p98 = np.percentile(rgb, (2, 98))
rgb = np.clip((rgb - p2) / (p98 - p2), 0, 1)

plt.figure(figsize=(6, 6))
plt.imshow(rgb)
plt.title(f'EuroSAT MSI RGB (B4-B3-B2) - Class: {sample["label"]}')
plt.axis('off')
plt.show()
```

## 2. EuroSAT SAR (Sentinel-1 Radar)
SAR data uses radar backscatter (VV and VH polarizations). Clay can handle these by padding or using specific band embeddings.

```
ds_sar = load_dataset('blanchon/EuroSAT_SAR', split='train', streaming=True)
sample_sar = next(iter(ds_sar))
img_sar = np.array(sample_sar['image'])
print(f'SAR Shape: {img_sar.shape}')
```

```
# Visualize SAR (VV Polarization)
vv = img_sar[0, :, :].astype(float)
p2, p98 = np.percentile(vv, (2, 98))
vv = np.clip((vv - p2) / (p98 - p2), 0, 1)

plt.figure(figsize=(6, 6))
plt.imshow(vv, cmap='gray')
plt.title('EuroSAT SAR (VV Polarization)')
plt.axis('off')
plt.show()
```
